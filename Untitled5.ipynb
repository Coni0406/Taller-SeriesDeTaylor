```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la función de aproximación de coseno usando una serie de Taylor
def cos_aprox(x: float, order: int) -> float:
    y = 1
    if order > 1:
        y -= (x**2) / 2
    if order > 3:
        y += (x**4) / 24
    if order > 5:
        y -= (x**6) / 720
    return y

# Rango de valores de x
x = np.linspace(-10, 10, 400)

# Graficar la función original de coseno
plt.figure(figsize=(10, 10))
plt.plot(x, np.cos(x), label=r'$\cos(x)$')

# Aproximaciones de la serie de Taylor para diferentes órdenes
for n in [2, 4, 6]:
    yi = []  # Almacena los valores de y para cada valor de x
    for xi in x:
        yi.append(cos_aprox(xi, n))
    plt.plot(x, yi, linestyle="--", label=f'Polinomio de Taylor de orden {n}')

# Configuración de la gráfica
plt.title("Aproximaciones de la función Coseno")
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-6, 6)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la función original g(x)
def g(x):
    return 1 / (1 - x)

# Definición de la función de aproximación usando la serie de Taylor acumulativa
def taylor_aprox(x: float, order: int) -> float:
    y = 1  # Empieza con el término constante
    if order > 1:
        y += x
    if order > 2:
        y += x**2
    if order > 3:
        y += x**3
    if order > 4:
        y += x**4
    if order > 5:
        y += x**5
    return y

# Rango de valores de x (excluyendo x=1 para evitar la singularidad)
x = np.linspace(-3, 2.5, 400)

# Configuración de la gráfica
plt.figure(figsize=(12, 8))

# Graficar la función original g(x)
plt.plot(x, g(x), color='red', label=r'$g(x) = \frac{1}{1 - x}$')

# Graficar los diferentes órdenes de la serie de Taylor
for n in range(1, 7):  # Desde el término constante hasta el término de grado 5
    yi = [taylor_aprox(xi, n) for xi in x]
    plt.plot(x, yi, linestyle="--", label=f'Polinomio de Taylor de orden {n}')

# Etiquetas y configuración final
plt.title("Aproximaciones de la función g(x) = 1 / (1 - x) con la Serie de Taylor")
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-10, 10)  # Ajuste para alejar la vista
plt.xlim(-3, 2.5)  # Ajuste del rango en x para ver mejor el comportamiento
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la función original g(x) = ln(x)
def g(x):
    return np.log(x)

# Definición de la función de aproximación acumulativa
def taylor_approximation(x, order):
    y = (x - 1)  # Empieza con el primer término
    if order > 1:
        y -= (1/2) * (x - 1)**2
    if order > 2:
        y += (1/3) * (x - 1)**3
    return y

# Rango de valores de x (excluyendo x <= 0 para evitar valores indefinidos en ln(x))
x = np.linspace(0.1, 10, 400)

# Configuración de la gráfica
plt.figure(figsize=(12, 8))

# Graficar la función original g(x) = ln(x)
plt.plot(x, g(x), color='red', label=r'$g(x) = \ln(x)$')

# Graficar los diferentes órdenes de la serie de Taylor
for n in range(1, 4):  # Desde el primer término hasta el tercer término
    yi = [taylor_approximation(xi, n) for xi in x]
    plt.plot(x, yi, linestyle="--", label=f'Polinomio de Taylor de orden {n}')

# Etiquetas y configuración final
plt.title("Aproximaciones del Polinomio f(x) = (x - 1) - 1/2(x - 1)^2 + 1/3(x - 1)^3 junto a g(x) = ln(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-5, 5)
plt.xlim(0, 10)
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
